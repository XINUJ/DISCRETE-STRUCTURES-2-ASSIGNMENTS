$\textbf{GRAPH}$

- A graph $G$ consists of two finite sets: a nonempty set $V(G)$ of vertices and a set $E(G)$ of edges, where each edge is associated with a set consisting of either one or two vertices called its endpoints. The correspondence from edges to endpoints is called the edge-endpoint function.

- An edge with just one endpoint is called a loop, and two or more distinct edges with the same set of endpoints are said to be parallel. An edge is said to connect its endpoints; two vertices that are connected by an edge are called adjacent; and a vertex that is an endpoint of a loop is said to be adjacent to itself.

- An edge is said to be incident on each of its endpoints, and two edges incident on the same endpoint are called adjacent. A vertex on which no edges are incident is called isolated.

If edge e is associated with the pair (v, w) of vertices, then e is said to be the (directed) edge from v to w.

$\textbf{SIMPLE GRAPH}$

- A simple graph is a graph that does not have any loops or parallel edges. In a simple
graph, an edge with endpoints $v$ and $w$ is denoted $\{v, w\}$.


Example:

Draw all simple graphs with the four vertices $\{v_1, v_2, v_3\}$ and two edges $e_1,e_2$. 

Solution:

There are $\begin{pmatrix}vertices \\ edges \end{pmatrix} = \frac{vertices!}{(edges!)(vertices-edges)!} = \begin{pmatrix}3 \\ 2 \end{pmatrix} =  \frac{3!}{(2!)(3-2)!} = 3$ ways to draw the simple graph. 

The entire subset is defined as $\{v_1, v_2\}, \{v_1, v_3\}, \{v_2, v_3\}$

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a list of all possible edges
edges = [
    [('v1', 'v2'), ('v1', 'v3')],
    [('v1', 'v2'), ('v2', 'v3')],
    [('v1', 'v3'), ('v2', 'v3')]
]

# Plot each graph
for idx, edge_list in enumerate(edges, start=1):
    # Create an empty graph
    G = nx.Graph()
    
    # Add nodes
    G.add_nodes_from(['v1', 'v2', 'v3'])
    
    # Add edges
    G.add_edges_from(edge_list)
    
    # Draw the graph
    plt.subplot(1, len(edges), idx)
    nx.draw(G, with_labels=True, node_color='skyblue', node_size=1000, font_size=12, font_weight='bold')
    plt.title(f"Graph {idx}")

plt.tight_layout()
plt.show()
```

$\textbf{BIPARTITE GRAPH}$

- Let $m$ and $n$ be positive integers. A complete bipartite graph on $(m, n)$ vertices, denoted $K_{m,n}$, is a simple graph with distinct vertices $v_1, v_2, . . . , v_m$ and $w_1, w_2, . . . , w_n$ that satisfies the following properties: $\forall i,k = 1,2,...,m$ and $\forall j,l = 1,2,...,n$

1. There is an edge from each vertex $v_i$ to each vertex $w_j$.
2. There is no edge from any vertex $v_i$ to any other vertex $v_k$.
3. There is no edge from any vertex $w_j$ to any other vertex $w_l$.

Example:

In the bipartite graph created, there are two distinct sets of nodes:

    Set 1: Nodes labeled 'A', 'B', 'C', and 'D'.
    Set 2: Nodes labeled 1, 2, and 3.

These two sets of nodes are disjoint, meaning that no edge connects nodes within the same set. This property is a characteristic feature of bipartite graphs.

The edges of the graph connect nodes from Set 1 to nodes from Set 2. Specifically, the following edges are present:

    ('A', 1)
    ('B', 1)
    ('B', 2)
    ('C', 2)
    ('D', 2)
    ('D', 3)

These edges represent relationships between entities in the two sets. For example, the edge ('A', 1) indicates that node 'A' is connected to node 1, and so on.

When visualized, the bipartite layout algorithm positions the nodes in such a way that nodes from Set 1 are placed on one side of the visualization, and nodes from Set 2 are placed on the other side. This layout helps to emphasize the bipartite structure of the graph.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a bipartite graph
B = nx.Graph()

# Add nodes with the node attribute "bipartite"
B.add_nodes_from(['A', 'B', 'C', 'D'], bipartite=0)  # Set of nodes on one side
B.add_nodes_from([1, 2, 3], bipartite=1)              # Set of nodes on the other side

# Add edges between nodes of different sets
B.add_edges_from([('A', 1), ('B', 1), ('B', 2), ('C', 2), ('D', 2), ('D', 3)])

# Plot the bipartite graph
pos = nx.bipartite_layout(B, ['A', 'B', 'C', 'D'])
nx.draw(B, pos=pos, with_labels=True, node_color='skyblue', node_size=1000, font_size=12, font_weight='bold')
plt.title("Bipartite Graph")
plt.show()
```

$\textbf{SUBGRAPH}$

A graph $H$ is said to be a subgraph of a graph $G$ if, and only if, every vertex in $H$ is also a vertex in $G$,every edge in $H$ is also an edge in $G$,and every edge in $H$ has the same endpoints as it has in $G$.

Example of a Subgraph: 

Create a subgraph H from an original graph G. 

G contains nodes [1, 2, 3, 4], and the edges (1, 2), (2, 3), (3, 4), (4, 1), and (2, 4).

In this example, the subgraph H contains only the nodes [1, 2, 3] and the edges between them from the original graph G.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph
G = nx.Graph()

# Add nodes and edges to the graph
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (2, 4)])

# Create a subgraph containing a subset of nodes and edges from the original graph
nodes_of_interest = [1, 2, 3]  # Nodes of interest
H = G.subgraph(nodes_of_interest)

# Plot the original graph
plt.figure(figsize=(8, 4))
plt.subplot(121)
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1000, font_size=12, font_weight='bold')
plt.title("Original Graph")

# Plot the subgraph
plt.subplot(122)
nx.draw(H, with_labels=True, node_color='lightgreen', node_size=1000, font_size=12, font_weight='bold')
plt.title("Subgraph")

plt.tight_layout()
plt.show()
```

$\textbf{HANDSHAKE THEOREM}$

The handshake theorem states that if $G$ is any graph, then the sum of the degrees of all the vertices of $G$ equals twice the number of edges of $G$. Specifically, if the vertices of $G$ are $v_1,v_2,...,v_n, where $n$ is a non-negative integer, then the total degree of $G = deg(v_1) + deg(v_2) +...+ deg(v_n) = 2$, which is the number of edges of $G.$

The Handshaken theorem can be used to determine if a Graph exists because the total of degree should always be even.

Example of Handshake Theorem:

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a simple graph
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)])

# Calculate the sum of degrees of all vertices
sum_of_degrees = sum(dict(G.degree()).values())

# Calculate the number of edges
num_edges = G.number_of_edges()

# Verify the Handshake Theorem
assert 2 * num_edges == sum_of_degrees, "Handshake Theorem is not satisfied"

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1000, font_size=12)
plt.title("Graph demonstrating Handshake Theorem")
plt.show()

print("Sum of degrees of all vertices:", sum_of_degrees)
print("Number of edges:", num_edges)
print("According to Handshake Theorem, 2 * num_edges should be equal to sum_of_degrees")
```

$\textbf{TRAVELLING IN A GRAPH}$

Travelling in a graph is accomplished by moving from one vertex to another along a sequence of adjaceny edges.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a simple graph
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 4), (2, 5), (3, 6), (3, 7)])

# Function for breadth-first search traversal
def bfs_traversal(graph, start):
    visited = set()
    queue = [start]
    while queue:
        vertex = queue.pop(0)
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            queue.extend(set(graph[vertex]) - visited)

# Perform BFS traversal starting from vertex 1
print("BFS traversal from vertex 1:")
bfs_traversal(G, 1)

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1000, font_size=12)
plt.title("Graph with BFS traversal from vertex 1")
plt.show()
```

$\textbf{EULER'S CIRCUIT}$

- Let $G$ be a graph, a $\textbf{Euler's circuit}$ that contains every vertex and every edge of G.
- The $\textbf{Euler's circuit}$ for $G$ is a sequence of adjacent vertices, and edges in $G$ that has at least one edge; starts and ends with the same vertex. Uses every vertex of G at least once, and uses every edge of $G$ $\textbf{"EXACTLY ONCE"}$.
- If $G$ is connected, and the degree of every vertex of $G$ is a positive even integer, then G has a $\textbf{Euler's circuit}$.
- $G$ has a $\textbf{Euler's circuit} \iff$ $G$ is connected and every vertex of $G$ has positive even degree.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create an Eulerian circuit graph
G = nx.Graph()

# Add edges to form an Eulerian circuit
edges = [(0, 1), (1, 2), (2, 3), (3, 0)]
G.add_edges_from(edges)

# Plot the graph
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1000, font_size=12, font_weight='bold')
plt.title("Eulerian Circuit Graph")
plt.show()
```

To prove that the graph created is an Eulerian circuit, it needs to satisfy two conditions:

Case 1: The graph is connected

Every vertex in the graph is reachable from every other vertex. Since the graph consists of a single loop with four vertices, it is indeed connected.

Case 2: Every vertex has even degree

In an Eulerian circuit, every vertex must have an even degree. In the graph, each vertex is connected to exactly two other vertices, so the degree of each vertex is 2, which is even.

Thus, the graph is an Eulerian Circuit.

$\textbf{EULER'S TRAIL}$

- Let $G$ be a graph, and $v$ and $w$ be distinct vertices in $G$, then a $\textbf{Euler's Trail}$ from $v$ to $w$ is a sequence of adjacent edges and vertices that starts at $v$ and ends at $w$, passes through every vertex of $G$ at least once, and traverses every edge of $G$ $\textbf{"EXACTLY ONCE"}$.
- If $v$ and $w$ are two distinct vertices of $G$, then there is a $\textbf{Euler's path}$ from $v$ to $w \iff G$ is connected; $v$ and $w$ have odd degree, and all other vertices of $G$ has a positive even degree. 

```python
import matplotlib.pyplot as plt
import networkx as nx
import random

def generate_eulerian_trail(num_nodes):
    # Create a graph
    G = nx.MultiGraph()

    # Add edges to form a trail
    for i in range(num_nodes - 1):
        G.add_edge(i, i + 1)

    # Make sure the last node connects to the first node
    G.add_edge(num_nodes - 1, 0)

    # Add random edges to ensure all nodes have even degree except two
    for i in range(num_nodes):
        while G.degree(i) % 2 != 0:
            j = random.randint(0, num_nodes - 1)
            if i != j and not G.has_edge(i, j):
                G.add_edge(i, j)

    return G

# Generate a random Eulerian trail with 8 nodes
num_nodes = 8
G = generate_eulerian_trail(num_nodes)

# Draw the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=12, font_weight='bold', arrows=True)

plt.title("Eulerian Trail Graph")
plt.show()
```

To prove that the graph we generated represents an Eulerian trail, it needs to satisfy two conditions:

    Case 1: The graph is connected
    
    Every vertex in the graph is reachable from every other vertex. In this case, the graph is created by connecting each node sequentially, ensuring that there is a path from every node to the next. 

    Case 2: Exactly two vertices have odd degrees, and all other vertices have even degrees
    
     In an Eulerian trail, all vertices except for the starting and ending vertices must have even degree. This property is ensured while creating the graph by adding extra edges randomly to make sure every node has an even degree, except for the first and last nodes.

Since the graph satisfies both conditions, it represents an Eulerian trail. An Eulerian trail is a path in a graph that visits every edge exactly once. In this case, the trail forms a loop, starting and ending at the same node, which is characteristic of an Eulerian trail. Thus, the graph we generated indeed represents an Eulerian trail.

```python
import matplotlib.pyplot as plt
import networkx as nx
import random

def nearest_neighbor(G, start):
    current = start
    unvisited = set(G.nodes())
    unvisited.remove(start)
    path = [start]
    
    while unvisited:
        next_node = min(unvisited, key=lambda x: G[current][x]['weight'])
        unvisited.remove(next_node)
        path.append(next_node)
        current = next_node
    
    path.append(start)  # Complete the circuit by returning to the start
    return path

def generate_random_graph(num_nodes):
    G = nx.complete_graph(num_nodes)
    
    for u, v in G.edges():
        if u != v:
            G.edges[u, v]['weight'] = random.randint(1, 10)
    
    return G

# Generate a random graph with 8 nodes
num_nodes = 8
G = generate_random_graph(num_nodes)

# Choose a random starting node
start_node = random.choice(list(G.nodes()))

# Find Hamiltonian circuit using Nearest Neighbor algorithm
hamiltonian_circuit = nearest_neighbor(G, start_node)

# Plot the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=12, font_weight='bold', arrows=True)

# Highlight the Hamiltonian circuit
nx.draw_networkx_edges(G, pos, edgelist=[(hamiltonian_circuit[i], hamiltonian_circuit[i+1]) for i in range(len(hamiltonian_circuit)-1)], edge_color='red', width=2)

plt.title("Hamiltonian Circuit")
plt.show()
```

Proving that the circuit generated is a Hamiltonian circuit involves demonstrating that it visits every vertex exactly once and returns to the starting vertex.

Let's analyze the properties of the circuit generated:

    Case 1: Visits every vertex exactly once: The Nearest Neighbor algorithm iteratively selects the nearest unvisited vertex from the current vertex until all vertices are visited 
    
    Since it never revisits a previously visited vertex (except for returning to the starting vertex at the end), and it ensures that all vertices are visited before completing the circuit, every vertex is visited exactly once.

    Case 2: Returns to the starting vertex
    
    The algorithm ensures that the circuit completes by returning to the starting vertex after visiting all other vertices. This ensures that the circuit forms a closed loop, satisfying the definition of a Hamiltonian circuit.

Thus, based on these observations, we can conclude that the circuit generated by the Nearest Neighbor algorithm indeed represents a Hamiltonian circuit.
