$\textbf{1. Kruskal's Algorithm}$

Kruskal's algorithm is an algorithm for finding the minimum spanning tree of an undirected edge-weighted graph. If the graph is connected, it finds a minimum spanning tree. It is a greedy algorithm that that makes use of the fact that the edges of a minimum spanning tree must form a subset of the edges of any other spanning tree.

Step 1: Find all minimum edge without forming a cycle.

$I - A$ (weight of 5)

$A - C$ (weight of 10)

$C - E$ (weight of 15)

$C - B$ (weight of 20)

$B - E$ (weight of 25): This forms a cycle with $C - B$ and $C - E$, so we skip this.

$C - D$ (weight of 30)

$E - F$ (weight of 35)

$I - E$ (weight of 40): This forms a cycle with $I - A$, $A - C$ and $C - E$, so we skip this.

$E - G$ (weight of 45)

Total cost: 160

Since we only have 8 vertices we stop here because of absolute rule spanning trees such that, $|e| = |v| - 1$.

$\textbf{Thus, this spanning tree has the lowest-cost edge, proven by Kruskal's Algorithm.}$

```python

import networkx as nx
import matplotlib.pyplot as plt

# Create the graph
G = nx.Graph()
G.add_nodes_from(["A","B","C","D","E","F","G","I"])
edgelist = [("A","C"),("A","I"),("C","B"),("C","E"),("C","D"),("B","E"),("I","E"),("E","F"),("F","G"),("G","E")]
G.add_edges_from(edgelist)

# Define the path you provided
path = [("I", "A"),("A", "C"), ("C", "E"),("B", "C"),("D", "C"), ("F", "E"),("G", "E")]

# Generate the minimum spanning tree with the edges of the path highlighted
MST = nx.Graph()
for edge in path:
    MST.add_edge(*edge)

# Plot the graph with highlighted edges
plt.figure(figsize=(10, 8))
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={("A","C"):10,("A","I"):5,("C","B"):20,("C","E"):15,("C","D"):30,("B","E"):25,("I","E"):40,("E","F"):35,("F","G"):50,("G","E"):45},
    font_color='red'
)
nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edges(MST, pos, edge_color='b', width=2)  # Add MST edges in blue
plt.title("Kruskal's Algorithm")
plt.show()
```

$\textbf{2. Prim's Algorithm}$

Prim's algorithm is a greedy algorithm that finds a minimum spanning tree for a weighted undirected graph. This means it finds a subset of the edges that forms a tree that includes every vertex, where the total weight of all the edges in the tree is minimized.

Step 1: From the graph, select the minimum cost edge.

That would be $I \rightarrow A$, which has a weight of 5.

Step 2: Then, select all the other edges with the minimum cost from the graph. However, it should be connected to the previously selected vertices and should not form a cycle, if not it would not be a spanning tree.


$I - A$ (weight of 5)

$A - C$ (weight of 10)

$C - E$ (weight of 15)

$C - B$ (weight of 20)

$B - E$ (weight of 25): This forms a cycle with $C - B$ and $C - E$, so we skip this.

$C - D$ (weight of 30)

$E - F$ (weight of 35)

$I - E$ (weight of 40): This forms a cycle with $I - A$, $A - C$ and $C - E$, so we skip this.

$E - G$ (weight of 45)

Total cost: 160

$\textbf{Thus, this spanning tree has the lowest-cost edge, proven by Prim's Algorithm.}$

```python

import networkx as nx
import matplotlib.pyplot as plt

# Create the graph
G = nx.Graph()
G.add_nodes_from(["A","B","C","D","E","F","G","I"])
edgelist = [("A","C"),("A","I"),("C","B"),("C","E"),("C","D"),("B","E"),("I","E"),("E","F"),("F","G"),("G","E")]
G.add_edges_from(edgelist)

# Define the path you provided
path = [("I", "A"),("A", "C"), ("C", "E"),("B", "C"),("D", "C"), ("F", "E"),("G", "E")]

# Generate the minimum spanning tree with the edges of the path highlighted
MST = nx.Graph()
for edge in path:
    MST.add_edge(*edge)

# Plot the graph with highlighted edges
plt.figure(figsize=(10, 8))
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={("A","C"):10,("A","I"):5,("C","B"):20,("C","E"):15,("C","D"):30,("B","E"):25,("I","E"):40,("E","F"):35,("F","G"):50,("G","E"):45},
    font_color='red'
)
nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edges(MST, pos, edge_color='b', width=2)  # Add MST edges in blue
plt.title("Prim's Algorithm")
plt.show()
```

$\textbf{3. Djikstra's Shortest Path Algorithm}$

Dijkstra's Algorithm is used for solving single-source shortest path problems for directed or undirected paths. Single-source means that one vertex is chosen to be the start (which is called the "source node"), and the algorithm will find the shortest path from that vertex to all other vertices. This algorithm uses the weights of the edges to find the path that minimizes the total distance (weight) between the source node and all other vertices.

Suppose that our source is $I(G)$ and destination is $F(G)$.

Step 1: Find all possible routes.

Case 1: $I \rightarrow E \rightarrow F$

Total distance: 75

Case 2: $I \rightarrow E \rightarrow G \rightarrow F$

Total distance: 135

Case 3: $I \rightarrow A \rightarrow C \rightarrow E \rightarrow F$

Total distance: 65

Case 4: $I \rightarrow A \rightarrow C \rightarrow E \rightarrow G \rightarrow F$

Total distance: 125

Case 5: $I \rightarrow A \rightarrow C \rightarrow B \rightarrow E \rightarrow  F$

Total distance: 95

Case 6: $I \rightarrow A \rightarrow C \rightarrow B \rightarrow E \rightarrow G \rightarrow F$

Total distance: 155

Step 2: Select shortest path from all possible routes.

$\textbf{Thus, the shortest route is the route in Case 3 which has a total distance of 65:}$

 $I \rightarrow A \rightarrow C \rightarrow E \rightarrow F$     
 
 5 + 10 + 15 + 35 = 65

```python

import networkx as nx
import matplotlib.pyplot as plt

# Create the graph
G = nx.Graph()
G.add_nodes_from(["A","B","C","D","E","F","G","I"])
edgelist = [("A","C"),("A","I"),("C","B"),("C","E"),("C","D"),("B","E"),("I","E"),("E","F"),("F","G"),("G","E")]
G.add_edges_from(edgelist)

# Define the path you provided
path = [("I", "A"),("A", "C"), ("C", "E"),("E", "F")]

# Generate the minimum spanning tree with the edges of the path highlighted
MST = nx.Graph()
for edge in path:
    MST.add_edge(*edge)

# Plot the graph with highlighted edges
plt.figure(figsize=(10, 8))
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={("A","C"):10,("A","I"):5,("C","B"):20,("C","E"):15,("C","D"):30,("B","E"):25,("I","E"):40,("E","F"):35,("F","G"):50,("G","E"):45},
    font_color='red'
)
nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edges(MST, pos, edge_color='b', width=2)  # Add MST edges in blue
plt.title("Djikstra's Shortest Path Algorithm")
plt.show()
```
